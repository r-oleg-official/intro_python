### Дана функция 
$$f(x) = 0,6x^3 + 5,5x^2 + 10x - 5 \tag{1}$$

1. Определить корни

2. Найти интервалы, на которых функция возрастает

3. Найти интервалы, на которых функция убывает

4. Построить график

5. Вычислить вершину

6. Определить промежутки, на котором f > 0

7. Определить промежутки, на котором f < 0

```python
a, b, c, d = [0.6, 5.5, 10, -5]
```

### 1. Определение корней.

```python
from sympy import *
# delta = (-4)*b**3*d + b**2*c**2 - 4*a*c**3 + 18*a*b*c*d - 27*a**2*d**2
x = symbols("x")
expr = a*x**3 + b*x**2 + c*x + d
expr_roots = solve(expr)
x1, x2, x3 = [round(expr_roots[0], 3), round(expr_roots[1], 3), round(expr_roots[2], 3)]
print(f'x1 = {x1}, x2 = {x2}, x3 = {x3}.')
```

### 2. Найти интервалы, на которых функция возрастает.

Для нахождения участка возрастания функции используется производная функции `f'(x)`. Если `f'(x) > 0` на этом интервале - то функция возрастает.

Для определения участка возрастания нужно найти экстремум функции - максимум. Для этого нужно найти корни выражения (3):

$$3ax^2 + 2bx + c = 0 \tag{2}$$
$$3*0,6x^2 + 2*5,5x + 10 = 0 <=> 1,8x^2 + 11x + 10 = 0 \tag{3}$$


```python
from math import sqrt
a1, b1, c1 = [1.8, 11, 10]

discrim = b1**2 - 4*a1*c1
x1_1 = (-b1 - sqrt(discrim)) / (2*a1)
y1_1 = a*x1_1**3 + b*x1_1**2 + c*x1_1 + d
x2_1 = (-b1 + sqrt(discrim)) / (2*a1)
y2_1 = a*x2_1**3 + b*x2_1**2 + c*x2_1 + d

infinity = b'\xe2\x88\x9e'.decode('utf-8')
print(f'Функция возрастает на участке \u2013 {infinity} \u00F7 {round(x1_1, 3), round(y1_1, 3)}.')
print(f'Функция возрастает на участке {round(x2_1, 3), round(y2_1, 3)} \u00F7 {infinity}.')
```

### 3. Найти интервалы, на которых функция убывает

Для нахождения участка убывания функции используется производная функции `f'(x)`. Если `f'(x) < 0` на этом интервале - то функция убывает.

Для определения участка возрастания нужно найти экстремум функции - максимум. Для этого нужно найти корни выражения (3).

```python
from math import sqrt
a1, b1, c1 = [1.8, 11, 10]

discrim = b1**2 - 4*a1*c1
x1_1 = (-b1 - sqrt(discrim)) / (2*a1)
y1_1 = a*x1_1**3 + b*x1_1**2 + c*x1_1 + d
x2_1 = (-b1 + sqrt(discrim)) / (2*a1)
y2_1 = a*x2_1**3 + b*x2_1**2 + c*x2_1 + d

print(f'Функция убывает на участке {round(x1_1, 3), round(y1_1, 3)} \u00F7 {round(x2_1, 3), round(y2_1, 3)}.')
```

### 4. Построить график


```python
import matplotlib.pyplot as plt

dec = [round(i*0.3, 3) for i in range(1, 6)]
x = []
for i in range(-8, 1):
    for item in dec:
        x_add = round(i + item, 1)
        if x_add < 1: x.append(x_add)
y1  = [round(a*i**3 + b*i**2 + c*i - d, 3) for i in x]
y2  = [round(3*a*i**2 + 2*b*i + c, 3) for i in x]

fig, ax = plt.subplots()

ax.set_title("f(x) = 0.6x^3 + 5.5x^2 + 10x - 5 \n f'(x) = 1.8x^2 + 11x + 10")
ax.grid()
ax = plt.gca() 
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.plot(x, y1, 'g^')
plt.plot(x, y1)
plt.plot(x, y2, 'b^')
plt.plot(x, y2)

fig.set_figwidth(20)
fig.set_figheight(10)

plt.show()
```

### 5. Вычислить вершину

Чтобы найти вершины (точки локальных минимумов и максимумов) кубической параболы, необходимо найти её производную, приравнять её к нулю и затем вычислить `(x, y)`.

```python
from math import sqrt
a1, b1, c1 = [1.8, 11, 10]

discrim = b1**2 - 4*a1*c1
x1_1 = (-b1 - sqrt(discrim)) / (2*a1)
y1_1 = a*x1_1**3 + b*x1_1**2 + c*x1_1 + d
x2_1 = (-b1 + sqrt(discrim)) / (2*a1)
y2_1 = a*x2_1**3 + b*x2_1**2 + c*x2_1 + d

print(f'Вершины кубической параболы: {round(x1_1, 3), round(y1_1, 3)} и {round(x2_1, 3), round(y2_1, 3)}.')
```
