### Дана функция f(x) = 5x^2 + 10x - 30

1. Определить корни

2. Найти интервалы, на которых функция возрастает

3. Найти интервалы, на которых функция убывает

4. Построить график

5. Вычислить вершину

6. Определить промежутки, на котором f > 0

7. Определить промежутки, на котором f < 0

```python
a, b, c = [5, 10, -30]
```

### 1. Определение корней.

```python
from math import sqrt

discrim = b**2 - 4*a*c

if discrim < 0:
    print("Корней нет.")
if discrim == 0:
    x = -(b / 2*a)
else:
    x1 = (-b + sqrt(discrim)) / (2*a)
    x2 = (-b - sqrt(discrim)) / (2*a)

print(f'X1 ={round(x1, 3)}')
print(f'X2 ={round(x2, 3)}')
```

### 2. Найти интервалы, на которых функция возрастает.

Для нахождения участка возрастания/убывания функции используется производная функции f'(x).

Для ax^2 + bx + c, производная:

    f'(x) = nax^(n -1) + bx^(n - 1).

Для данной функции первая производная:

    2*5*x + 10
    10x +10

Для поиска участка возрастания функции решить неравенство:

    10x + 10 > 0
    10x > -10
    x > -1

Функция возрастает при `x > -1`.

```python
x_up = -1
y_up = a*x_up**2 + 10*x_up - 30
infinity = b'\xe2\x88\x9e'.decode('utf-8')
print(f'Функция возрастает от {x_up, y_up} - {infinity}')
```

### 3. Найти интервалы, на которых функция убывает


Для нахождения участка возрастания/убывания функции используется производная функции f'(x).

Для ax^2 + bx + c, производная:

    f'(x) = nax^(n -1) + bx^(n - 1).

Для данной функции первая производная:

    10x +10

Для поиска участка возрастания функции решить неравенство:

    10x + 10 < 0
    x < -1

Функция убывает при `x < -1`.

```python
x_down = -1
y_down = a*x_down**2 + 10*x_down - 30
infinity = b'\xe2\x88\x9e'.decode('utf-8')
print(f'Функция убывает от {infinity} - {x_down, y_down}')
```

### 4. Построить график


```python
import matplotlib.pyplot as plt

x, y  = [], []
for i in range(-7, 6):
    x.append(i)
    y.append(5*i**2 + 10*i - 30)

fig, ax = plt.subplots()

ax.set_title("Graph 5x^2 + 10x - 30")
ax.grid()
ax = plt.gca() 
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.plot(x, y, 'g^')
plt.plot(x, y)
fig.set_figwidth(20)
fig.set_figheight(10)

plt.show()
```
